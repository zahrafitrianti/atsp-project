```python
import nltk 
from nltk.tokenize import word_tokenize
tria = []
substr = "permission.".lower()
file = open("nvpZdL4O.txt",newline='')
result = file.read()
words = word_tokenize(result)
for i in words:
    tria.append(i)
    print(i)
```

```python
import re
permi = []
for i in range(1, len(tria)):
    permissions = re.findall(r'permission.*', tria[i])
    if permissions:
        permi.append(tria[i])
        print(permi)
    else:
        print("sorry")
        print()
```
