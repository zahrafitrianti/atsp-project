```python
import pandas as pd
import numpy as np
```

```python
def averageovercolumns():
    liste = []
    final = []
    data_c = pd.read_csv('final_scores_normalized.csv')
    # total final frame
    final1 = data_c.rename(index=str, columns={"combined score": "combinedscore"})
    # top 5 frame
    final = final1.sort_values('combinedscore', ascending=False).head(5)[['combinedscore', 'app_name']]
    allscore = final1['combinedscore']
    scorelist = final['combinedscore']
    scorelist = scorelist.tolist()
    allscore = allscore.tolist()
    #final = data_c.rename(index=str, columns={"combined features": "combinedfeatures"})
    applist = final['app_name']
    applist =applist.tolist()
    
    score_pp = np.asarray(scorelist)
    app_pp = np.asarray(applist)
    flat_score = [sublist for sublist in score_pp ]
    flat_appname = [sublist for sublist in app_pp ]
    liste.append(flat_score)
    liste.append(flat_appname)
    return flat_score,flat_appname,allscore
```

```python
flat_score,flat_appname,allscore = averageovercolumns()
```
